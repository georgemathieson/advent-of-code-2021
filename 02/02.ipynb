```
using System.IO;

var input = File.ReadLines("input");
```

```
var horizontalPosition = 0;
var depth = 0;

foreach (var line in input)
{
    var command = line.Split(' ').First();
    var unit = Int32.Parse(line.Split(' ').Last());

    if (command == "forward") horizontalPosition += unit;
    if (command == "backwards") horizontalPosition -= unit;
    if (command == "up") depth -= unit;
    if (command == "down") depth += unit;
}

Console.WriteLine($"Part 1: {horizontalPosition * depth}");
```

```
var horizontalPosition = 0;
var depth = 0;
var aim = 0;

foreach (var line in input)
{
    var command = line.Split(' ').First();
    var unit = Int32.Parse(line.Split(' ').Last());

    if (command == "forward")
    {
        horizontalPosition += unit;
        depth += (aim * unit);
    }

    if (command == "backwards")
    {
        horizontalPosition -= unit;
    }
    
    if (command == "up")
    {
        aim -= unit;
    }

    if (command == "down") 
    {
        aim += unit;
    }
}

Console.WriteLine($"Part 2: {horizontalPosition * depth}");
```
