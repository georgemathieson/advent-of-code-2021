```
using System.IO;

var input = File.ReadLines("input")
    .Select(line => Int32.Parse(line))
    .ToArray();
```

```
int previous = default;
int count = default;

foreach (var line in input)
{
    if (line != input.First() && line > previous)
    {
        count++;
    }

    previous = line;
}

Console.WriteLine($"Part 1: {count}");
```

```
int previous = default;
int count = default;

for (int i = 0; i < input.Count() - 2; i++)
{
    var sum = input[i] + input[i+1] + input[i+2];

    if (i != 0 && sum > previous)
    {
        count++;
    }

    previous = sum;
}

Console.WriteLine($"Part 2: {count}");
```
